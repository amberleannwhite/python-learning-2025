```python
amber_profile = {
    "name": "Amber",
    "role": "Data Engineer (in training)",
    "learning": ["Python", "GitHub", "SQL"],
    "focus": "Systems Thinking",
    "superpower": "Architectural Clarity"
}
print(amber_profile)
```

```python
print(amber_profile["role"])
print(amber_profile["learning"][0])  # the first item in the learning list
```

```python
amber_profile["learning"].append("Linux")
print(amber_profile["learning"])
```

```python
amber_profile["learning"].remove("SQL")
print(amber_profile["learning"])
```

```python
amber_profile["certifications"] = ("IBM Data Engineering Essentials")
print(amber_profile)
```

```python
amber_profile["progress"] = "Module 2: Data Structures in Python" 
print(amber_profile["progress"])
```

```python
amber_profile["superpower"] = "Optimizing Complex Systems with Structure"
print(amber_profile["superpower"])
```

```python
intro = f"""
Hi, I'm **{amber_profile['name']}** - a {amber_profile['role']}!
I'm currently learning: {", ".join(amber_profile['learning'])}
My focus area is: {amber_profile['focus']}
And my not-so-secret superpower is: {amber_profile['superpower']}
"""
print(intro)
```

```python
amber_profile["projects"] = [
    {
        "name": "Python Foundations",
        "type": "learning",
        "status": "in progress",
        "tech": ["Python", "Jupyter", "GitHub"],
    },
    {
        "name": "Data Engineering Essentials",
        "type": "certification",
        "status": "completed",
        "provider": "IBM / Coursera",
        "focus": ["pipelines", "databases", "ETL basics"],
    },
    {
        "name": "Portfolio & GitHub Setup",
        "type": "portfolio",
        "status": "in progress",
        "tech": ["Git", "GitHub", "Markdown"],
    },
]
```

```python
print(amber_profile["projects"][0]["name"])
print(amber_profile["projects"][1]["status"])
```

```python
for project in amber_profile["projects"]:
    name = project["name"]
    status = project["status"]
    ptype = project["type"].title()
    print(f"- {name}  [{ptype}]: {status}")
```

```python
print("\nActive focus areas:")
for project in amber_profile["projects"]:
    if project["status"] == "in progress":
        print(" •", project["name"])
```

```python
amber_profile["timeline"] = {
    "pre_2024": "Systems thinking, creative production, cross-department communication",
    "2024_q4": "Transition into data engineering, IBM Data Engineering Essentials",
    "2025_q1": "Deepening Python + SQL, building portfolio projects, ready for career gear shift",
}
for period, focus in amber_profile["timeline"].items():
    print(period, "→", focus)
```
