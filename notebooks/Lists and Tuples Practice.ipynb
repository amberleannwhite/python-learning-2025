```python
# Lists & Tuples Practice — Python Learning 2025

Small practice notebook reinforcing:
- List creation, indexing & `.append()`
- Tuple immutability
- Converting strings ↔ lists using `.split()` + `.join()`

Learning mindset:
> “I learn fast and apply even faster.”
```

```python
tools = ["Python", "GitHub", "SQL"]
tools.append("Pandas")

print(tools)
# Output: ['Python', 'GitHub', 'SQL', 'Pandas']
```

```python
tech = ["Python", "SQL", "GitHub", "Linux", "Airflow"]
tech.append("Pandas")

print("Count:", len(tech))
print("First:", tech[0])
print("Last:", tech[-1])
```

```python
projects = ["pipeline", "storage", "analytics"]

print("First project:", projects[0])
print("Last project:", projects[-1])  # Negative indexing works too!

projects.append("monitoring")
print("Updated list:", projects)

cloud_stack = ("AWS", "Azure", "GCP")
print("Public cloud options:", cloud_stack[1:])
```

```python
skills = [ "Python", "SQL", "GitHub"]
skills.append("Apache Spark")
print(skills[2:4])
print(skills)
```

```python
skills = [ "Python", "SQL", "GitHub"]
skills.append("Apache Spark")
print(skills[2:4])
print(" & " .join(skills))
```

```python
skills = [ "Python", "SQL", "GitHub"]
skills.append("Apache Spark")
print(skills[2:4])
print("Skills:", ", " .join(skills))
```

```python
skills = [ "Python", "SQL", "GitHub"]
skills.insert(1, "Linux") #insert at specific index

print(", ".join(skills))
```

```python
skills = ["Python", "Linux", "SQL", "GitHub"]
skills[1] = "Linux (Intermediate)"
print("Skills:", ", ".join(skills))
```

```python
clouds = ["AWS", "Azure", "GCP", "OCI"]
print("First cloud:", clouds[0])
print("Last cloud:", clouds[-1])
print("Second-to-last cloud:", clouds[-2])
```

```python
skills = ["Python", "GitHub", "Linux", "SQL"]
name = "Amber"
focus = ["Python", "GitHub"]

main_skills = skills[0:2]  # Python, GitHub
focus_str = " and ".join(focus)

skills_str = ", ".join(main_skills)
all_skills = ", ".join(skills[:-1]) + ", and " + skills[-1]

print(f"Hi, my name is {name} and my skills are {skills_str} with a current focus in {focus_str}.")
print(f"My full skill set includes {all_skills}.")
```
