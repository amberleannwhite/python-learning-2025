```python
#GET "Cat Facts" via API

import requests
import pandas as pd

#start with an empty list
facts = []

#collect 5 cat facts in a list
for _ in range(5):
    response = requests.get("https://catfact.ninja/fact")
    data = response.json()
    facts.append(data["fact"])

#make a DataFrame
df_facts = pd.DataFrame({"Cat Fact": facts})

#print the results
print("\nüê±Cat Facts Table:\n")
print(df_facts)

#save to a CSV
df_facts.to_csv("cat_facts.csv", index=False)
print("\nüíæ Saved 'cat_facts.csv' successfully!")
```

```python
#Reusable template #1

import requests
import pandas as pd

def fetch_cat_facts(n=5, save_path=None):
    """
    Fetch n random cat facts from the API and return a DataFrame.
    If save_path is given, also save the results to a CSV file.
    """

    facts = []
    for _ in range(n):
        response = requests.get("https://catfact.ninja/fact", timeout=5)
        response.raise_for_status()
        data = response.json()
        facts.append(data.get("fact", "No fact field returned"))

    df = pd.DataFrame({"Cat Fact": facts})

    if save_path:
        df.to_csv(save_path, index=False)

    return(df)
```
