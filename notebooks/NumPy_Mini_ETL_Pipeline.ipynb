```python
#THE CUTEST PET CHALLENGE! - Mini ETL SORTING SIM
#begin
import numpy as np

#1D array - "Cutest Pets Voting Sim"
cuteness_level = np.array([100, 3000, 500, 60])
print("â‹†ğŸ¾Â° Cuteness Level (Popular Vote):", cuteness_level)
print("Dimensions:", cuteness_level.ndim)
print("Shape:", cuteness_level.shape)
print("Size:", cuteness_level.size)

#2D array (table) - "Cutest Pets, scores from contest judges"
scores = np.array([[430, 820],[3000, 3000],[90, 91]])

print("\nâ‹†ğŸ¾Â°Cuteness Score (Judges' Scores):\n", scores)
print("Dimensions:", scores.ndim)
print("Shape:", scores.shape)
print("Size:", scores.size)
```

```python
#weight each column, ex: judges score 40%, popular vote 60%
weights = np.array([0.4, 0.6])

#matrix multiplication
final_scores = scores.dot(weights.T)
#structural flatten
final_scores = final_scores.flatten()

print("\nFinal Weighted Scores:")
print(final_scores)
```

```python
#Data Pulls
print("Cutest Pet:", final_scores.max())
print("Average Cuteness:", final_scores.mean())

#Identify index placements of highest score
top_index = final_scores.argmax()
print("Top Pet - row index:", top_index)
print("Their Original Cuteness:", scores[top_index])
```

```python
#Name Pets, Identify Winner
pets = np.array(["Roger", "Cecelia", "Barton von Louisville III"])
print("Cutest Pet:", pets[top_index])
```

```python
#Final Vote tallies
print("\nAll final scores:")
for i, score in enumerate(final_scores):
    print(pets[i], "â†’", score)

print("\nTop Pets(>= 500):")
top_pets = final_scores >= 500
print(pets[top_pets])

print("\nPets Sorted by Rank:")
ranked_idx = final_scores.argsort()[::-1]
for idx in ranked_idx:
    print(final_scores[idx], "â†’", pets[idx])
```

```python
#Export .CSV Data with Pandas
import pandas as pd

#Combine pet names and scores into a Data Frame
results_df = pd.DataFrame({"Pet": pets, "Final Score": final_scores})

#save to CSV
results_df.to_csv("cutest_pet_results.csv", index=False)

print("CSV Saved! ğŸ“‚âœ¨")
results_df
```

```python
#Import/LOAD .CSV with Pandas
import pandas as pd

#Load back the CSV
loaded_df = pd.read_csv("cutest_pet_results.csv")

print("ğŸ“¥ Loaded Data Frame:")
print(loaded_df)
```

```python
#Now normalize scores on a 0-100 scale

loaded_df["Normalized_Score"] = (loaded_df["Final Score"] / loaded_df["Final Score"].max() * 100 ) .round(2)

print("\nAfter normalization:")
print(loaded_df)
```

```python
#Sort by scores, descending
sorted_df = loaded_df.sort_values(by="Normalized_Score", ascending=False)

#Export Final polished report
sorted_df.to_csv("Cutest_Pet_Report_Final.csv)", index=False)

print("\nâœ¨ Final report Saved! ğŸ“‚âœ¨\n")
print(sorted_df)
```
