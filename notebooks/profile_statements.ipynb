```python
def learning_statement(language, years):
    if years ==0:
        return(f"I am just starting my journey learning {language}")
    elif years == 1:
        return(f"I have been learning {language} for {years} year")
    else:
        return(f"I have been using {language} for {years} years")
print(learning_statement("Python", 1))
print(learning_statement("Python", 0))
print(learning_statement("Python", 5))
```

```python
def profile_summary(name, role, language, years):
    if years ==0:
        return(f"Hi, I'm {name}, and I just started my {role} role. I have just started learning {language}")
    elif years == 1:
        return(f"Hi, I'm {name}, and I just started my {role} role. I have been learning {language} for {years} year")
    else:
        return(f"Hi, I'm {name}, and I just started my {role} role. I have been learning {language} for {years} years")
print(profile_summary("Amber", "Data Engineering", "Python", 0))
```
