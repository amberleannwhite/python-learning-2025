```python
#GET "Cat Facts" 

import requests
import pandas as pd

#start with an empty list
facts = []

#collect 5 cat facts in a list
for _ in range(5):
    response = requests.get("https://catfact.ninja/fact")
    data = response.json()
    facts.append(data["fact"])

#make a DataFrame
df_facts = pd.DataFrame({"Cat Fact": facts})

#print the results
print("\nðŸ±Cat Facts Table:\n")
print(df_facts)

#save to a CSV
df_facts.to_csv("cat_facts.csv", index=False)
print("\nðŸ’¾ Saved 'cat_facts.csv' successfully!")
```

```python
#THE CUTEST PET CHALLENGE! - Mini ETL SORTING SIM
#begin
import numpy as np

#1D array - "Cutest Pets Voting Sim"
cuteness_level = np.array([100, 3000, 500, 60])
print("â‹†ðŸ¾Â° Cuteness Level (Popular Vote):", cuteness_level)
print("Dimensions:", cuteness_level.ndim)
print("Shape:", cuteness_level.shape)
print("Size:", cuteness_level.size)

#2D array (table) - "Cutest Pets, scores from contest judges"
scores = np.array([[430, 820],[3000, 3000],[90, 91]])

print("\nâ‹†ðŸ¾Â°Cuteness Score (Judges' Scores):\n", scores)
print("Dimensions:", scores.ndim)
print("Shape:", scores.shape)
print("Size:", scores.size)
```

```python
#weight each column, ex: judges score 40%, popular vote 60%
weights = np.array([0.4, 0.6])

#matrix multiplication
final_scores = scores.dot(weights.T)
#structural flatten
final_scores = final_scores.flatten()

print("\nFinal Weighted Scores:")
print(final_scores)
```

```python
#Data Pulls
print("Cutest Pet:", final_scores.max())
print("Average Cuteness:", final_scores.mean())

#Identify index placements of highest score
top_index = final_scores.argmax()
print("Top Pet - row index:", top_index)
print("Their Original Cuteness:", scores[top_index])
```

```python
#Name Pets, Identify Winner
pets = np.array(["Roger", "Cecelia", "Barton von Louisville III"])
print("Cutest Pet:", pets[top_index])
```

```python
#Final Vote tallies
print("\nAll final scores:")
for i, score in enumerate(final_scores):
    print(pets[i], "â†’", score)

print("\nTop Pets(>= 500):")
top_pets = final_scores >= 500
print(pets[top_pets])

print("\nPets Sorted by Rank:")
ranked_idx = final_scores.argsort()[::-1]
for idx in ranked_idx:
    print(final_scores[idx], "â†’", pets[idx])
```

```python
#Export .CSV Data with Pandas
import pandas as pd

#Combine pet names and scores into a Data Frame
results_df = pd.DataFrame({"Pet": pets, "Final Score": final_scores})

#save to CSV
results_df.to_csv("cutest_pet_results.csv", index=False)

print("CSV Saved! ðŸ“‚âœ¨")
results_df
```

```python
#Import/LOAD .CSV with Pandas
import pandas as pd

#Load back the CSV
loaded_df = pd.read_csv("cutest_pet_results.csv")

print("ðŸ“¥ Loaded Data Frame:")
print(loaded_df)
```

```python
#Now normalize scores on a 0-100 scale

loaded_df["Normalized_Score"] = (loaded_df["Final Score"] / loaded_df["Final Score"].max() * 100 ) .round(2)

print("\nAfter normalization:")
print(loaded_df)
```

```python
#Sort by scores, descending
sorted_df = loaded_df.sort_values(by="Normalized_Score", ascending=False)

#Export Final polished report
sorted_df.to_csv("Cutest_Pet_Report_Final.csv)", index=False)

print("\nâœ¨ Final report Saved! ðŸ“‚âœ¨\n")
print(sorted_df)
```

```python
#Cat Facts Console!
#define the action fetch_cat_facts

import requests
import pandas as pd
def fetch_cat_facts(n=1, filename=None):
    """Get n cat facts from the API and (optionally) save to a CSV."""
    facts = []
    for _ in range(n):
        response = requests.get("https://catfact.ninja/fact")
        data = response.json()
        facts.append(data["fact"])

    df = pd.DataFrame({"Cat Fact": facts})

    if filename:
        df.to_csv(filename, index=False)

    return df
```

```python
#Define the Console
       
def cat_console():
    print("Welcome to the Cat Data Console!")

    while True:
        print("\nWhat would you like to do?")
        print("1. Get a random Cat Fact")
        print("2. Get several Cat Facts and save to .csv")
        print("3. Show our 'Cutest Pet Results' from voting sim")
        print("4. Quit")

        choice = input("Enter 1-4:").strip()

        if choice == "1":
            df = fetch_cat_facts(1)
            print("\nYour Random Cat Fact:\n", df["Cat Fact"].iloc[0])
        elif choice == "2":
            n = input("How many facts? (default 5): ").strip()
            n = int(n) if n else 5
            df = fetch_cat_facts(n, "cat_facts.csv")
            print(f"\nSaved {len(df)} facts to 'cat_facts.csv'")
        elif choice == "3":
            print("\nCutest Pet Results:\n")
            print(sorted_df)
        elif choice == "4":
            print("Goodbye from the Cat Console!")
            break
        else:
            print("Please enter 1, 2, 3, or 4.")
```
